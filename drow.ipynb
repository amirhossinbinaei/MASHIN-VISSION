````python
import cv2
import mediapipe as mp
import numpy
import math
from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
import bidi

def put_text_farsi(frame, text, org, color):
    frame_pillow = Image.fromarray(frame)
    text_reshap = arabic_reshaper.reshape(text)
    text_farsi = bidi.get_display(text_reshap)
    frame_font = ImageFont.truetype('amir.ttf', 20)
    pen = ImageDraw.Draw(frame_pillow)
    pen.text(org, text_farsi, color, frame_font)
    return numpy.array(frame_pillow)

# ุฑุงูโุงูุฏุงุฒ Mediapipe
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(min_detection_confidence=0.7, min_tracking_confidence=0.7,max_num_hands=2)

# ูุนุงูโุณุงุฒ ูุจโฺฉู
cap = cv2.VideoCapture(0)

# ููุงุฏุฑ ุงููู
brush_size = 10
slider_x = 580
slider_y = 150
slider_w = 20
slider_h = 250
slider_val = 100
jahat_dast = None
erasing_mode = False
was_in_eraser_zone = False



# ูุงุญูโูุง
trigger_zone = (540, 10, 620, 50) # ูุงุญู ูพุงฺฉโฺฉู (ูุฑูุฒ)
black_zone   = (495, 10, 535, 50) #     ูุงุญู ุฑูฺฏ ุณุงู
white_zone   = (450, 10, 490, 50) #ูุงุญู ุฑูฺฏ ุณูุฏ
green_zone   = (405, 10, 445, 50) #ูุงุญู ุฑูฺฏ ุณุจุฒ
pink_zone    = (360, 10, 400, 50) #    ูุงุญู ุฑูฺฏ ุตูุฑุช
red_zone     = (315, 10, 355, 50) #ูุงุญู ุฑูฺฏ ูุฑูุฒ
purpel_zone  = (270, 10, 310, 50) #ูุงุญู ุฑูฺฏ ูุฑูุฒ
blue_zone    = (225, 10, 265, 50) #      ูุงุญู ุฑูฺฏ ุขุจ
yellow_zone  = (180, 10, 220, 50) #      ูุงุญู ุฑูฺฏ ุฒุฑุฏ
Brown_zone   = (135, 10, 175, 50) # ูุงุญู ุฑูฺฏ ูููู ุง
Orange_zone  = ( 90, 10, 130, 50) #   ูุงุญู ุฑูฺฏ ูุงุฑูุฌ
eraser_zone  = (400, 10, 520, 50) #  ูุงุญู ูพุงฺฉ ฺฉู ูุญู
erasing_mode = False              # ูพุงฺฉ ฺฉู ูุญู ุบุฑ ูุนุงู ุงุณุช

# ูุณุฑ ุงูฺฏุดุช ู ูุถุนุช ุฑุณู
all_trails = []
current_trail = []
redo_stack = []    # redo

trail_color=(0,0,0) # ุฑูฺฏ ูพุด ูุฑุถ ููุงุด ุ ุณุงู ุงุณุช
drawing = False     #      ุนููุงุช ุฑุณู ุบุฑ ูุนุงู ุงุณุช

#ุฑุงู ุงูุฏุงุฒ ูุจฺฉู
while cap.isOpened():
    ret, frame = cap.read()
    frame_slider = frame.copy()
    if not ret:
        break
    frame_slider = cv2.flip(frame_slider, 1)
    frame        = cv2.flip(frame       ,  1)
    h, w, _ = frame.shape
    x_index, y_index = 0, 0  # ููุฏุงุฑ ุงููู

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(rgb_frame)
    if result.multi_hand_landmarks:
        for hand_landmarks , d2 in zip(result.multi_hand_landmarks,result.multi_handedness):
            landmarks = hand_landmarks.landmark
            jahat_dast = d2.classification[0].label
            x_index  = int(landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP].x * w)   # ูุฎุชุตุงุช ูุญูุฑ x ุงูฺฏุดุช ุงุดุงุฑู
            y_index  = int(landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP].y * h)   # ูุฎุชุตุงุช ูุญูุฑ y ุงูฺฏุดุช ุงุดุงุฑู

            x_THUMB = int(landmarks[mp_hands.HandLandmark.THUMB_TIP].x * w)           # ูุฎุชุตุงุช ูุญูุฑ x ุงูฺฏุดุช ุดุตุช
            y_THUMB = int(landmarks[mp_hands.HandLandmark.THUMB_TIP].y * h)           # ูุฎุชุตุงุช ูุญูุฑ y ุงูฺฏุดุช ุดุตุช

            distance = math.hypot(x_index - x_THUMB, y_index - y_THUMB) # ูุญุงุณุจู ูุงุตูู ุงูฺฏุดุช ุงุดุงุฑู ู ุดุตุช ุจุฑ ุญุณุช ุชุนุฏุงุฏ ูพฺฉุณู


            if distance < 40 and  jahat_dast=='Right':
                if erasing_mode:
                    # ูพุงฺฉ ฺฉุฑุฏู ูุณุฑูุง ุฏุฑ ูุฒุฏฺฉ ุงูฺฏุดุช
                    updated_trails = []
                    for trail, color, size in all_trails:
                        split_trail = []
                        current_segment = []

                        for point in trail:
                            dist = math.hypot(point[0] - x_index, point[1] - y_index)
                            if dist > 20:
                                current_segment.append(point)
                            else:
                                if len(current_segment) > 1:
                                    split_trail.append(current_segment)
                                current_segment = []

                        # ุงุถุงูู ฺฉุฑุฏู ุขุฎุฑู ูุทุนู ุงฺฏุฑ ูุนุชุจุฑ ุจูุฏ
                        if len(current_segment) > 1:
                            split_trail.append(current_segment)

                        for segment in split_trail:
                            updated_trails.append((segment, color, size))

                    all_trails = updated_trails

                    continue  # ฺูู ุฏุงุฑู ูพุงฺฉ ูโฺฉููุ ุฏฺฏู ุงุฏุงูู ููโุฏู ุจุฑุง ุฑุณู
                else:
                    if not drawing:
                        drawing = True
                        current_trail = []
                    current_trail.append((x_index, y_index))

            else:
                if drawing:
                    drawing = False
                    if current_trail:
                        all_trails.append((current_trail, trail_color, brush_size))  # ุฐุฎุฑู ููุฑุงู ุฑูฺฏ
                        current_trail = []



    # ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ูพุงฺฉโฺฉู ุดุฏุ ูพุงฺฉ ูโฺฉูุฏ ูุณุฑูุง ู ุฑูฺฏ ุณุงู ุฑุง ุฎุงููุด ูโฺฉูุฏ
    x1, y1, x2, y2 = trigger_zone
    if  x1 <= x_index <= x2 and y1 <= y_index <= y2 and jahat_dast=='Right':
        all_trails = []
        current_trail = []
        erasing_mode = False

    # ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ุณุงู ุดุฏุ ุญุงูุช ุฑูฺฏ ุณุงู ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    bx1, by1, bx2, by2 = black_zone
    if jahat_dast == 'Left' and bx1 <= x_index <= bx2 and by1 <= y_index <= by2:
        trail_color = (0,0,0)
        erasing_mode = False

      #ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ุณูุฏ ุดุฏุ ุญุงูุช ุฑูฺฏ ุณูุฏ ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    wx1, wy1, wx2, wy2 = white_zone
    if jahat_dast == 'Left' and wx1 <= x_index <= wx2 and wy1 <= y_index <= wy2:
        trail_color = (255, 255, 255)
        erasing_mode = False

    #ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ุณุจุฒ ุดุฏุ ุญุงูุช ุฑูฺฏ ุณุจุฒ ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    gx1, gy1, gx2, gy2 = green_zone
    if jahat_dast == 'Left' and  gx1 <= x_index <= gx2 and gy1 <= y_index <= gy2:
        trail_color = (0, 255, 0)
        erasing_mode = False

    #ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ุตูุฑุช ุดุฏุ ุญุงูุช ุฑูฺฏ ุตูุฑุช ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    px1, py1, px2, py2 = pink_zone
    if jahat_dast == 'Left' and px1 <= x_index <= px2 and py1 <= y_index <= py2:
        trail_color = (147, 20, 255)
        erasing_mode = False

    # ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ูุฑูุฒ ุดุฏุ ุญุงูุช ุฑูฺฏ ูุฑูุฒ ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    rx1, ry1, rx2, ry2 = red_zone
    if jahat_dast == 'Left' and rx1 <= x_index <= rx2 and ry1 <= y_index <= ry2:
        trail_color = (0, 0, 255)
        erasing_mode = False

    # ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ุจููุด ุดุฏุ ุญุงูุช ุฑูฺฏ ุจููุด ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    pux1, puy1, pux2, puy2 = purpel_zone
    if jahat_dast == 'Left' and pux1 <= x_index <= pux2 and puy1 <= y_index <= puy2:
        trail_color = (128, 0, 128)
        erasing_mode = False

    # ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ุขุจ ุดุฏุ ุญุงูุช ุฑูฺฏ ุขุจ ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    bluex1, bluey1, bluex2, bluey2 = blue_zone
    if jahat_dast == 'Left' and bluex1 <= x_index <= bluex2 and bluey1 <= y_index <= bluey2:
            trail_color = (255, 0, 0)
            erasing_mode = False

    # ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ุฒุฑุฏ ุดุฏุ ุญุงูุช ุฑูฺฏ ุฒุฑุฏ ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    yx1, yy1, yx2, yy2 = yellow_zone
    if jahat_dast == 'Left' and yx1 <= x_index <= yx2 and yy1 <= y_index <= yy2:
        trail_color = (0, 255, 255)
        erasing_mode = False

    # ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ูููู ุง ุดุฏุ ุญุงูุช ุฑูฺฏ ูููู ุง ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    Bx1, By1, Bx2, By2 = Brown_zone
    if jahat_dast == 'Left' and Bx1 <= x_index <= Bx2 and By1 <= y_index <= By2:
            trail_color = (19, 69, 139)
            erasing_mode = False

    # ุงฺฏุฑ ุงูฺฏุดุช ูุงุฑุฏ ูุงุญู ุฑูฺฏ ูุงุฑูุฌ ุดุฏุ ุญุงูุช ุฑูฺฏ ูุงุฑูุฌ ูุนุงู ุดูุฏ ู ุจุฑุง ููุดู ุจูุงูุฏ
    nx1, ny1, nx2, ny2 = Orange_zone
    if jahat_dast == 'Left' and nx1 <= x_index <= nx2 and ny1 <= y_index <= ny2:
        trail_color = (0, 165, 255)
        erasing_mode = False
    # ูพุงฺฉ ฺฉู ูุญู
    ex1, ey1, ex2, ey2 = eraser_zone
    in_eraser_zone = ex1 <= x_index <= ex2 and ey1 <= y_index <= ey2 and  jahat_dast=='Right'
    if in_eraser_zone and not was_in_eraser_zone:
        erasing_mode = not erasing_mode  # ุชุบุฑ ูุถุนุช ูพุงฺฉโฺฉู (toggle)
    was_in_eraser_zone = in_eraser_zone

    # ุจูโุฑูุฒุฑุณุงู ุณุงุฒ ููู
    brush_size = max(1, int((slider_val / slider_h) * 50))

    # ุฑุณู ููุงุฑ ูุบุฒุงู ุฏุฑ ูพูุฌุฑูโ ฺฉูุชุฑู
    cv2.rectangle(frame_slider, (slider_x, slider_y), (slider_x + slider_w, slider_y + slider_h), (200, 200, 200), -1)
    cv2.rectangle(frame_slider, (slider_x, slider_y), (slider_x + slider_w, slider_y + slider_val),  trail_color, -1)
    cv2.putText(frame_slider, f'Brush Size: {brush_size}', (slider_x-85, slider_y - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2)


    # ุจุฑุฑุณ ุจุฑุฎูุฑุฏ ุจุง ุงุณูุงุฏุฑ
   # if jahat_dast == 'Left':
    if jahat_dast == 'Left' and slider_x <= x_index <= slider_x + slider_w and \
            slider_y <= y_index <= slider_y + slider_h:
        slider_val = y_index - slider_y


    # ุฑุณู ูุณุฑูุง ูุจู
    for trail, color, size in all_trails:
        for i in range(1, len(trail)):
             cv2.line(frame, trail[i - 1], trail[i], color, size)

    # ุฑุณู ูุณุฑ ูุนู
    for i in range(1, len(current_trail)):
         cv2.line(frame, current_trail[i - 1], current_trail[i], trail_color,  brush_size)

    # ุฑุณู ูุงุญูโูุง
    cv2.rectangle(frame, (trigger_zone[0], trigger_zone[1]),(trigger_zone[2],trigger_zone[3]), (0, 0, 255), 2)
    #frame = put_text_farsi(frame, "clear_all", (trigger_zone[0] + 10, trigger_zone[1] + 5), (0, 0, 0))
    cv2.putText(frame, "clear_all", (trigger_zone[0] + 1, trigger_zone[1] + 20),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2)

    cv2.rectangle(frame_slider, (black_zone[0] , black_zone[1]) , (black_zone[2], black_zone[3])  , (  0,   0,   0), -1)

    cv2.rectangle(frame_slider, (white_zone[0] , white_zone[1]) , (white_zone[2], white_zone[3])  , (255, 255, 255), -1)

    cv2.rectangle(frame_slider, (green_zone[0] , green_zone[1]) , (green_zone[2], green_zone[3])  , (  0, 255,   0), -1)

    cv2.rectangle(frame_slider, (pink_zone[0]  , pink_zone[1])  , (pink_zone[2] , pink_zone[3])   , (147,  20, 255), -1)

    cv2.rectangle(frame_slider, (red_zone[0]   , red_zone[1])   , (red_zone[2]  , red_zone[3])    , (  0,   0, 255), -1)

    cv2.rectangle(frame_slider, (purpel_zone[0], purpel_zone[1]), (purpel_zone[2], purpel_zone[3]), (128,   0, 128), -1)

    cv2.rectangle(frame_slider, (blue_zone[0]  , blue_zone[1])  , (blue_zone[2]  , blue_zone[3])  , (255,   0,   0), -1)

    cv2.rectangle(frame_slider, (yellow_zone[0], yellow_zone[1]), (yellow_zone[2], yellow_zone[3]), (  0, 255, 255), -1)

    cv2.rectangle(frame_slider, (Brown_zone[0] , Brown_zone[1]) , (Brown_zone[2] , Brown_zone[3]) , ( 19,  69, 139), -1)

    cv2.rectangle(frame_slider, (Orange_zone[0], Orange_zone[1]), (Orange_zone[2], Orange_zone[3]), (  0, 165, 255), -1)


    # ุงฺฏุฑ ูพุงฺฉโฺฉู ูุนุงููุ ูุณุชุทู ูพุฑ ุฑูฺฏ ุชุฑ (ูุฑูุฒ ูพุฑ) ุฑุณู ุจุดู
    if erasing_mode:
        cv2.rectangle(frame, (eraser_zone[0], eraser_zone[1]), (eraser_zone[2], eraser_zone[3]), (0, 0, 200), -1)
    else:
        cv2.rectangle(frame, (eraser_zone[0], eraser_zone[1]), (eraser_zone[2], eraser_zone[3]), (100, 100, 100), 2)
    # ููุดุชู ูุชู ุฑู ูุณุชุทูุ ุฏุฑ ูุฑ ุฏู ุญุงูุช
    cv2.putText(frame, "local_eraser", (eraser_zone[0] + 1, eraser_zone[1] + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255) if erasing_mode else (0, 0, 0), 2)



    if jahat_dast == 'Right':
     cv2.circle(frame, (x_index, y_index),  brush_size , trail_color, -1) # ููุทู ุงูฺฏุดุช

    if jahat_dast == 'Left':
         cv2.circle(frame_slider, (x_index, y_index), brush_size, trail_color, -1)  # ููุทู ุงูฺฏุดุช




         # ููุงุด ฺฉุงุฏุฑ ุฏูุฑ ุฑูฺฏ ูุนุงู
    # ุฏฺฉุดูุฑ ุฑูฺฏโูุง ุจู ููุงุญ ูุฑุจูุทู
    color_zone_map = {
        (0, 0, 0): black_zone,
        (255, 255, 255): white_zone,
        (0, 255, 0): green_zone,
        (147, 20, 255): pink_zone,
        (0, 0, 255): red_zone,
        (128, 0, 128): purpel_zone,
        (255, 0, 0): blue_zone,
        (0, 255, 255): yellow_zone,
        (19, 69, 139): Brown_zone,
        (0, 165, 255): Orange_zone
    }
    if trail_color in color_zone_map:
        zone = color_zone_map[trail_color]
        # ุฑุณู ฺฉุงุฏุฑ ุฏูุฑ ูุงุญู ุฑูฺฏ ูุนุงู
        cv2.rectangle(frame_slider, (zone[0], zone[1]), (zone[2], zone[3]), (0, 0, 0), 2)

    status_text = f"Color: {trail_color} | Size: {brush_size} | Mode: {'Erasing' if erasing_mode else 'Drawing'}"
    cv2.putText(frame_slider, status_text, (10, h - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (trail_color), 2)

    cv2.imshow("Control Panel", frame_slider)
    cv2.imshow("Air Drawing", frame)

    key = cv2.waitKey(1) & 0xFF



    # Undo
    if key == ord('z'):
        if all_trails:
            redo_stack.append(all_trails.pop())

    # Redo
    elif key == ord('y'):
        if redo_stack:
            all_trails.append(redo_stack.pop())


    elif key == ord('s'):  # Save
        cv2.imwrite('drawing.png', frame)
    elif key == ord('q') or cv2.getWindowProperty('Air Drawing', cv2.WND_PROP_VISIBLE) < 1:
        break

cap.release()
cv2.destroyAllWindows()
````
